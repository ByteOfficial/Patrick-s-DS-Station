```python
import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
tb = pd.read_csv('tb.csv')
print(tb.info())
tb.head()
```

```python
tb.columns
```

```python
tbmelt = tb.melt(id_vars = ['country','year'], 
        value_vars =['m04', 'm514', 'm014', 'm1524', 'm2534', 'm3544',
       'm4554', 'm5564', 'm65', 'mu', 'f04', 'f514', 'f014', 'f1524', 'f2534',
       'f3544', 'f4554', 'f5564', 'f65', 'fu'],
        var_name = 'sexage', value_name = 'cases'
       )
```

```python
tbmelt.info()
```

```python
tbmelt['year'] = pd.to_datetime(tbmelt['year'], errors='ignore' )
# use .dt 
tbmelt['year']  = tbmelt['year'].dt.year
tbmelt.head()
```

```python
tbmelt['sex'] = tbmelt['sexage'].str.slice(0,1)
tbmelt['age'] = tbmelt['sexage'].str.slice(1)
```

```python
tbmelt.head()
```

```python
tbmelt['age'] = tbmelt['age'].map({'04': '0-4', '514': '5-14', '1524': '15-24', '2534': '25-34', '3544': '35-44',
                                   '4554': '45-54', '5564': '55-64', '65': '65+', 'u': np.nan})
```

```python
final = tbmelt.dropna(subset = ['cases'])
```

```python
final.head()
```

```python
final.info()
```

```python
final = final.sort_values(['country','year','age','sex'])
```

```python
final.drop(['sexage'],axis =1, inplace = True)
```

```python
final.head(10)
```

```python
final.to_csv('finaltb.csv',index=False)
```
