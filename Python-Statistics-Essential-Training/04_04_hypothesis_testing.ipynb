Make simulation use area 0 1 4 5 which have the most cases of cholera and simulate each death randomly,proportionally to the population of each area.

```python
def simulate(n):
    return pd.DataFrame({'closest': np.random.choice([0,1,4,5], size = n, p = [0.65,0.15,0.10,0.10])})
# p is the proportiona of polulation in each area
```

```python
simulate(489).closest.value_counts()
```

```python
# extract the count for area zero, repeat the opeartion 10,000 time
sampling = pd.DataFrame({'counts':[simulate(489).closest.value_counts()[0] for i in range(10000)]})
```

```python
sampling.counts.hist(histtype = 'step')
```

Generated the distribution under the null hypothesis that the pumps have nothing to do with color and the death occur simply proportionally to population.

We can now compare the distribution with the observed number of 340 deaths in area zero. More precisely, we evaluate at what quantile we find 340 in this null hypothesis sampling distribution.

```python
scipy.stats.percentileofscore(sampling.counts, 340)
```

```python
# p value, the larger the P value, the more strongly we can reject the null hypothesis.
100- 98.17
```
